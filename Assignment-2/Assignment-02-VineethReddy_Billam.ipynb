## Problem: 2

Generate an array of length $3n$ , filled with the cyclic pattern 1, 2, 3. Where $n$ is a natural number.

```python
# Write your code here
def cyclic_pattern(n):
    l =[1,2,3] * n
    return (np.array(l))
print(cyclic_pattern(4))

```

## Problem: 3
Create an array of the first $n$ odd integers. $n$ is a natural number input from user.

```python
# Write your code here
def odd(n):
    l = np.arange(start=1,stop=n,step=  2)
    return (l)
print(odd(10))
```

## Problem: 4

Create a 10 x 10 arrays of zeros and then "frame" it with a border of ones.

```python
# Write your code here
u =np.zeros((10,10))   
u[0,:] = 1
u[-1,:] = 1
u[:,0] = 1
u[:,-1] = 1
    
print(u)    
```

## Problem: 5

Create an 8 x 8 array with a checkerboard pattern of zeros and ones.

```python
# Write your code here
u = np.zeros((8,8))
for i,j in np.ndindex(u.shape):
    if (i+j) % 2:
        u[i,j] = 0
    else:
        u[i,j] = 1
print(u)
```

## Problem: 7 
Create a function which creates an $n×n$ array with $(i , j)$ -entry equal to $i+ j$

```python
# Write your code here
def function1(n):
    arr = np.zeros((n,n))
    for i,j in np.ndindex(arr.shape):
        arr[i,j] = i+j
    return (arr)

print(function1(5))
```

# Problem: 8
Evaluate $cos$ and $sin$ on the interval $[0,1]$  with 1000 points and then stack the results into a tall array with rows being the $(cos(x),sin(x))$ entries.

```python
# Write your code here
x = np.linspace(0,1,100)
c = np.cos(x)
s = np.sin(x)

tall_array = np.stack((c,s),axis=1)
print(tall_array)
```

# Problem: 9

Create a random $3×5$ array using the `np.random.rand(3 ,5)` function and compute: the sum of all the entries, the sum of the rows, and the sum of the columns.

```python
# Write your code here

ar1 = np.random.rand(3,5)
total_sum = ar1.sum()
row_sum = ar1.sum(axis=1)
col_sum = ar1.sum(axis=0)
print("sum of all entries:", total_sum)
print("sum of rows:", row_sum)
print("sum of columns:", col_sum)
```

## Problem: 10

Create a random $5×5$ array using the function `np.random.rand(5, 5)`. Then sort the rows according to the second column.

```python
# Write your code here
ar2 = np.random.rand(5,5)
second_Col = ar2[:,1]
s = np.argsort(second_Col)
sorted_order = ar2[s]
print("sorted array:\n", sorted_order)   
```
